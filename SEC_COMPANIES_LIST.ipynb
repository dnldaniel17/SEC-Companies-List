```python
import numpy as np
import pandas as pd
import requests
from openpyxl import Workbook, load_workbook
from datetime import date 

# CIK (Unique company number), Ticker Symbol, and Company names are pulled from the SEC website (JSON file) to give 
# you an updated list of companies complying with SEC Filing regulations. 
# This code will pull the most recent companies lise from the SEC website

# STEP 1 - JSON to dict
ticker_url = r"https://www.sec.gov/files/company_tickers.json"
content = requests.get(ticker_url)
decoded_ticker_content = content.json() # JSON structure that will turn into a python dictionary.

# STEP 2 - Create Pandas dataframe  
ticker_content_df = pd.DataFrame(decoded_ticker_content).T
ticker_content_df.rename(columns = {'cik_str':'CIK','ticker':'Ticker','title':'Company Name'}, inplace = True)

# STEP 3 -Create excel spreadsheet
today_str = str(date.today()).replace('-', '')
wb = Workbook()
ws = wb.active
ws.title = "Data"
wbname = "Companies_list_{date}.xlsx".format(date = today_str)
wb.save(wbname)

# STEP 4 - Convert dataframe into spreadsheet
ticker_content_df.to_excel(wbname,index=False)
```
